```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import tree, metrics
from sklearn import model_selection
import graphviz
```

```python
data = pd.read_csv('/Users/nilesh/Downloads/car.data',
names=['buying','maint','doors','persons','lug_boot','safety','class'])
data.head(200)
```

```python
data.info()
```

```python
data['class'],class_names = pd.factorize(data['class'])
```

```python
print(class_names)
```

```python
print(data['class'].unique())
```

```python
data['buying'],_ = pd.factorize(data['buying'])
```

```python
data['maint'],_ = pd.factorize(data['maint'])
data['doors'],_ = pd.factorize(data['doors'])
data['persons'],_ = pd.factorize(data['persons'])
data['lug_boot'],_ = pd.factorize(data['lug_boot'])
data['safety'],_ = pd.factorize(data['safety'])
```

```python
data.head(1000)
```

```python
data.info()
```

```python
X = data.iloc[:,:-1]
```

```python
y = data.iloc[:,-1]
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33, random_state=0)
```

```python
dtree = tree.DecisionTreeClassifier(criterion='entropy', random_state=0)
```

```python
dtree.fit(X_train,y_train)
```

```python
y_pred = dtree.predict(X_test)
len(y_pred)
```

```python

count_misclassified = (y_test != y_pred).sum()
```

```python
print('Misclassified samples: {}'.format(count_misclassified))
```

```python
accuracy = metrics.accuracy_score(y_test, y_pred)
```

```python
print('Accuracy: {:.2f}'.format(accuracy))
```
